```python
from keras.datasets import cifar10
import matplotlib.pyplot as plt

# Lista de nomes para cada classe do dataset CIFAR10
cifar10_class_names = {
    0: "Plane",
    1: "Car",
    2: "Bird",
    3: "Cat",
    4: "Deer",
    5: "Dog",
    6: "Frog",
    7: "Horse",
    8: "Ship",
    9: "Truck"
}

# Carregar o dataset inteiro
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Varredura em cada imagem no dataset
for i in range(30): # mostrando apenas 30 imagens do dataset (que contém 60000 imagens, 6000 para cada uma das 10 classes)
  # Pegar uma imagem do dataset
  sample_image = x_train[i]
  # Pegar o id da classe associada à imagem
  image_class_number = y_train[i][0]
  # Procurar o nome da classe baseado no id da classe
  image_class_name = cifar10_class_names[image_class_number]

  # Plotar a imagem
  plt.imshow(sample_image)
  # Rotular a imagem
  plt.title(image_class_name)
  # Mostrar o plot na tela
  plt.show()
```
